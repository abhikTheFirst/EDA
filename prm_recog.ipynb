```python
import re
import numpy as np

# Load parameter file to parse atom type definitions
param_file_path = 'amoebabio18.prm'


# Function to parse the parameter file for atom type definitions
def parse_atom_types(param_file_path):
    atom_type_dict = {}
    with open(param_file_path, 'r') as file:
        for line in file:
            if line.startswith("atom"):
                parts = line.split()
                atom_type = int(parts[1])  # Atom type ID
                residue_name = parts[4:6]  
                atom_type_dict[atom_type] = residue_name
    return atom_type_dict

def find_residue_connections(residueTypes):
    np.unique(residueTypes, return_counts=True)




# Parse atom types from the provided parameter file

atom_types = parse_atom_types(param_file_path)
atom_types_list = list(atom_types.items())
atom_types_list

```
